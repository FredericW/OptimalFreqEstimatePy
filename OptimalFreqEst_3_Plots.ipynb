```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# set key parameters
d = 20
dist_type = "exp"
est_type = "sw"

filename = 'data/results_%s_%s_%d' %(est_type,dist_type,d)
temp = pd.read_csv(filename)
eps_grid = np.asarray(temp['eps_grid'])

aaa_dist = np.asarray(temp['aaa_dist'])
est_dist = np.asarray(temp['%s_dist'%(est_type)])

aaa_var = np.asarray(temp['aaa_var'])
est_var = np.asarray(temp['%s_var' %(est_type)])


plt.plot(eps_grid,aaa_var,label = 'aaa')
plt.plot(eps_grid, est_var,label = est_type)
plt.xlabel('epsilon')
plt.ylabel('variance')
plt.legend()
plt.grid()
plt.show()

plt.semilogy(eps_grid,aaa_dist,label = 'aaa w. %s'%(est_type))
plt.semilogy(eps_grid, est_dist,label = est_type)
plt.xlabel('epsilon')
plt.ylabel('Wass Distance')
plt.legend()
plt.grid()
plt.show()
```
