```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# set key parameters
d = 10 # num of bins in [low,high]
beta = 1
data_type = "GAUSSIAN"
est_type = "sw"
N=d*100000
est_repeat = 5
test_repeat = 30

filename = 'data/result_%s_%s_%d_(%d,%d).csv' %(data_type,est_type,d,
                                                est_repeat,test_repeat)
temp = pd.read_csv(filename)
eps_grid = np.asarray(temp['eps_grid'])
var_aaa = np.asarray(temp['var_aaa'])
var_est = np.asarray(temp['var_est'])
wass_aaa = np.asarray(temp['wass_aaa'])
wass_est = np.asarray(temp['wass_est'])

plt.figure(figsize=(5,3))
plt.plot(eps_grid,var_aaa,label = 'aaa w. %s'%(est_type),
         linewidth='2.0')
plt.plot(eps_grid, var_est,label = est_type,
         linewidth='2.0',
         linestyle='dashed')
plt.xlabel('epsilon')
plt.ylabel('variance')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(5,3))
plt.plot(eps_grid,wass_aaa,label = 'aaa w. %s'%(est_type),
         linewidth='2.0')
plt.plot(eps_grid, wass_est,label = est_type,
         linewidth='2.0',
         linestyle='dashed')
plt.xlabel('epsilon')
plt.ylabel('Wass Distance')
plt.legend()
plt.grid()
plt.show()
```
