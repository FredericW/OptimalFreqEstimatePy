```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# set key parameters
data_type="GAUSSIAN" 
d = 16
est_type="sw"
test_type='rounds'
repeat=100
portion = 0.2

filename = 'data/result_%s_%s_%d_%s(%d).csv' % (data_type,
                        est_type, d,test_type, repeat)
result = pd.read_csv(filename)
eps_grid = np.asarray(result['eps_grid'])
var_aaa = np.asarray(result['var_aaa'])
var_est = np.asarray(result['var_est'])
wass_aaa = np.asarray(result['wass_aaa'])
wass_est = np.asarray(result['wass_est'])

plt.figure(figsize=(5,3))
plt.plot(eps_grid,var_aaa,label = 'aaa w. %s'%(est_type),
        linewidth='2.0')
plt.plot(eps_grid, var_est,label = est_type,
        linewidth='2.0',
        linestyle='dashed')
plt.xlabel('epsilon')
plt.ylabel('MSE')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(5,3))
plt.semilogy(eps_grid,wass_aaa,label = 'aaa w. %s'%(est_type),
        linewidth='2.0')
plt.semilogy(eps_grid, wass_est,label = est_type,
        linewidth='2.0',
        linestyle='dashed')
plt.xlabel('epsilon')
plt.ylabel('Wass Distance')
plt.legend()
plt.grid()

filename = 'figure/fig_%s_%s_%d_%s(%.2f,%d).png' % (data_type,
                        est_type, d,test_type, portion, repeat)
plt.savefig(filename)
plt.show()
```
