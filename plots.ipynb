```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# set key parameters
data_type="GAUSSIAN" 
d = 8
est_type="grr"
test_type='rounds'
total_repeat=100

if test_type=='rounds':
    est_repeat=10
    filename = 'data/result_%s_%s_%d_(%d,%d).csv' %(data_type,est_type,d, est_repeat,total_repeat)
elif test_type=='portions':
    portion = 0.1
    filename = 'data/result_%s_%s_%d_(%.2f,%d)_%s.csv' % (data_type,
                                                 est_type, d, portion, total_repeat,test_type)
else:
    raise NotImplementedError('check your test_type status!')

result = pd.read_csv(filename)
eps_grid = np.asarray(result['eps_grid'])
var_aaa = np.asarray(result['var_aaa'])
var_est = np.asarray(result['var_est'])
wass_aaa = np.asarray(result['wass_aaa'])
wass_est = np.asarray(result['wass_est'])

plt.figure(figsize=(5,3))
plt.plot(eps_grid,var_aaa,label = 'aaa w. %s'%(est_type),
        linewidth='2.0')
plt.plot(eps_grid, var_est,label = est_type,
        linewidth='2.0',
        linestyle='dashed')
plt.xlabel('epsilon')
plt.ylabel('variance')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(5,3))
plt.semilogy(eps_grid,wass_aaa,label = 'aaa w. %s'%(est_type),
        linewidth='2.0')
plt.semilogy(eps_grid, wass_est,label = est_type,
        linewidth='2.0',
        linestyle='dashed')
plt.xlabel('epsilon')
plt.ylabel('Wass Distance')
plt.legend()
plt.grid()
plt.show()
```
