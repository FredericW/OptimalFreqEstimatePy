```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# set key parameters
data_type="GAUSSIAN" 
d = 16
est_type="sw"
test_type='rounds'
ratio=0
repeat=50

filename = 'data/result_%s_%s_%d(%.1f,%d).csv' % (data_type,
                        est_type, d, ratio, repeat)
result = pd.read_csv(filename)
eps_grid = np.asarray(result['eps_grid'])
var_aaa = np.asarray(result['mse_aaa'])
var_est = np.asarray(result['mse_est'])
wass_aaa = np.asarray(result['wass_aaa'])
wass_est = np.asarray(result['wass_est'])

plt.figure(figsize=(5,3))
plt.plot(eps_grid,var_aaa,label = 'aaa w. %s'%(est_type),
        linewidth='2.0')
plt.plot(eps_grid, var_est,label = est_type,
        linewidth='2.0',
        linestyle='dashed')
plt.xlabel('epsilon', fontsize=15)
plt.ylabel('MSE', fontsize=15)
plt.legend(fontsize=13)
plt.grid()
plt.show()

plt.figure(figsize=(5,3))
plt.semilogy(eps_grid,wass_aaa,label = 'aaa w. %s'%(est_type),
        linewidth='2.0')
plt.semilogy(eps_grid, wass_est,label = est_type,
        linewidth='2.0',
        linestyle='dashed')
plt.xlabel('epsilon', fontsize=15)
plt.ylabel('Wass Distance', fontsize=15)
plt.legend(fontsize=13)
plt.grid()

filename = 'figure/fig_%s_%s_%d(%d).png' % (data_type,
                        est_type, d, repeat)
plt.savefig(filename)
plt.show()
```
